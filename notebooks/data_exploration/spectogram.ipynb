```python
import mne
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir = "/pasteur/zeus/projets/p02/BCOM/ciprian_project/data_analyzed/non_normalized/data/WITHOUT_BADS/COVERT/BCOM_26_4_i_116-epo.fif"
```

```python
epos = mne.read_epochs(dir)
```

```python
epos = epos.crop(-0.2, 0.6)
```

```python
sample_1 = epos[0]
```

```python
for i in range(len(epos)):
    frequencies = np.array(range(1, 150, 2))  # Frequency range from 2 to 40 Hz, in steps of 2
    n_cycles = frequencies / 5

    powers = []
    # Number of cycles, adjust as needed
    power = mne.time_frequency.tfr_morlet(
    epos[i], 
    freqs=frequencies, 
    n_cycles=n_cycles, 
    use_fft=True, 
    return_itc=False,  # Only power, not inter-trial coherenc
    n_jobs=3)          # Number of parallel jobs
    powers.append(power)     
    power.plot(
        picks='MEG 022',  # Replace with specific channel name or type, e.g., 'MEG 1332'
        baseline=None,  # Baseline correction (optional)
        mode='logratio', # Baseline normalization method
        cmap='viridis',
        title="Spectrogram"
    )
```
