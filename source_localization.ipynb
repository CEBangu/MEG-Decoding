```python
import mne
import os
from datahandling import BcomMEG
```

```python
dir = '/Volumes/@neurospeech/PROJECTS/BCI/BCOM/DATA_ANALYZED/EVOKED/DATA/WITHOUT_BADS/COVERT'
participant = 'BCOM_01'
epo_files = []
subjects = [participant + '_2', participant + '_3', participant + '_4'] # focus on one because the head thing for the mutliple trials is not working atm. 
avoid_reading = False
avoid_producing = True
data = BcomMEG(dir=dir, subjects=subjects, avoid_reading=avoid_reading, avoid_producing=avoid_producing)
```

```python
# get all of the epochs into one object
epochs = []
for subject in data.data:
    for epoch in data.data[subject]:
        epochs.append(mne.concatenate_epochs([data.data[subject][epoch]]))

epochs = mne.concatenate_epochs(epochs, on_mismatch='ignore')
raw = mne.io.read_raw_fif(f'/Volumes/@neurospeech/PROJECTS/BCI/BCOM/DATA_ANALYZED/PREPROCESSED/{participant}/2/cleaned_raw.fif', preload=True)
```

```python
# noise convariance esitmation (quantify variability in the data)
noise_cov = mne.compute_covariance(epochs=epochs, tmin=-0.3, tmax=0.5)
```

```python
# set up forward model with template

# get subject directory
subject = 'fsaverage'
subjects_dir = str(mne.datasets.sample.data_path()) + '/subjects'

# create source space for cortex
src = mne.setup_source_space(subject, spacing='oct6', subjects_dir=subjects_dir,
                             add_dist=False)

# Create BEM (Boundary Element Method) model and solution
model = mne.make_bem_model(subject=subject, ico=4, subjects_dir=subjects_dir)
bem = mne.make_bem_solution(model)
```

```python
mne.gui.coregistration(subject='sample', subjects_dir=subjects_dir)
```

```python
# template transformation matrix
trans = '/Users/ciprianbangu/Cogmaster/M2 Internship/BCI code/co_registration/coreg_01.fif'
# trans = 'fsaverage'
fwd = mne.make_forward_solution(raw.info, trans=trans, src=src, 
                                bem=bem, eeg=False, meg=True)
```

```python
# compute the inverse operator
inverse_operator = mne.minimum_norm.make_inverse_operator(raw.info, fwd, noise_cov, 
                                                          loose=0.2, depth=0.8)
```

```python
# applying inverse operator
evoked = epochs.average()
snr = 3.0
lambda2 = 1.0/snr**2
stc = mne.minimum_norm.apply_inverse(evoked, inverse_operator, lambda2,
                                     method='dSPM', pick_ori=None)
```

```python
# Parcellation/Mapping
labels = mne.read_labels_from_annot(subject, parc='aparc', subjects_dir=subjects_dir)

# extract time-series for the labels
label_ts = mne.extract_label_time_course(stc, labels, src, mode='mean')
```

```python
stc.plot(subject=subject, subjects_dir=subjects_dir)
```
