```python
import mne
import os
import os.path as op
import pandas as pd
```

```python
original_dir_covert = "/Volumes/BCOM/BCOM/DATA_ANALYZED/EVOKED/DATA/WITHOUT_BADS/COVERT"
original_dir_overt = "/Volumes/BCOM/BCOM/DATA_ANALYZED/EVOKED/DATA/WITHOUT_BADS/OVERT"
```

```python
epoch_dataframe_original_covert = pd.DataFrame({"syllable": [],
                                "count": []})
epoch_dataframe_original_overt = pd.DataFrame({"syllable": [],
                                "count": []})
```

```python
for file in os.listdir(original_dir_covert):
    
    if "BCOM" in file:

        file_path = op.join(original_dir_covert, file)
        syllable = file[10:].split("-")[0]

        epochs = mne.read_epochs(file_path, verbose=False)
        count = len(epochs.events)

        if syllable not in epoch_dataframe_original_covert['syllable'].values:
            epoch_dataframe_original_covert = pd.concat([epoch_dataframe_original_covert, pd.DataFrame({"syllable": [syllable], "count": [count]})], ignore_index=True)
        else: 
            epoch_dataframe_original_covert.loc[epoch_dataframe_original_covert['syllable'] == syllable, 'count'] += count
```
